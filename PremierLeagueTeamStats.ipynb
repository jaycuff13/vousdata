```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#import CSV File 

matches = pd.read_csv("/Users/JordanCuff/Downloads/matches.csv")

```

```python
#Anaylsis data using .head() function

matches.head()
```

```python
#Getting total value counts of match results

matches['result'].value_counts()
```

```python
#Calculate total wins by team across all seasons 

matches['team'].value_counts() + matches[matches['result'] == 'W']['team'].value_counts(ascending=False)
```

```python
matches['venue'].value_counts()
```

```python
#Calculation total home losses by each team 

matches['team'].value_counts() + matches[matches['result'] == 'L'][matches['venue'] == 'Home']['team'].value_counts()
```

```python
#Calculation total home wins by each team 

matches['team'].value_counts() + matches[matches['result'] == 'W'][matches['venue'] == 'Home']['team'].value_counts()
```

```python
#Horizontal Bar graph that calculates total wins both home and away by each team 

((matches[matches['result'] == 'W'][matches['venue'] == 'Home']['team'].value_counts()) + 
(matches[matches['result'] == 'W'][matches['venue'] == 'Away']['team'].value_counts())).sort_values().plot(kind='barh')
```

```python
#Horizontal Bar graph that calculates total losses both home and away by each team 

((matches[matches['result'] == 'L'][matches['venue'] == 'Home']['team'].value_counts()) + 
(matches[matches['result'] == 'L'][matches['venue'] == 'Away']['team'].value_counts())).sort_values().plot(kind='barh')
```

```python
#Pie chart of w/ pct to show share of total home wins by each team 

((matches[matches['result'] == 'W'][matches['venue'] == 'Home']['team'].value_counts())).sort_values().plot(kind='pie', autopct='%1.1f%%')
```

```python
#Pie chart of w/ pct to show share of total wins by each team on Mondays 

((matches[matches['result'] == 'W'][matches['day'] == 'Mon']['team'].value_counts())).sort_values().plot(kind='pie', autopct='%1.1f%%')
```

```python
#Pie chart that shows gf for share percent in Arsenal overall wins 

((matches['gf'][matches['result'] == 'W'][matches['team'] == 'Arsenal'].value_counts())).sort_values().plot(kind='pie', autopct='%1.1f%%')
```
